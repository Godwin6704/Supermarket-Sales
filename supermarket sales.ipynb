### The line df['Total'] = df[' Total '].str.replace('$', '').astype('float') is used to clean and convert the "Total" column in a pandas DataFrame. It removes the dollar sign ($) from the values, ensuring they are properly formatted as numbers, and then converts the column from a string type to a floating-point number (float). The space in " Total " suggests a column name formatting issue, which should be corrected using df.columns = df.columns.str.strip().

```python
df['Total'] = df[' Total '].str.replace('$', '').astype('float')
```

### The line df['Tax 5%'] = df['Tax 5%'].str.rstrip('%').astype('float') / 100 is used to clean and convert the "Tax 5%" column from a percentage string to a decimal format. The str.rstrip('%') function removes the percentage symbol (%) from the values, and .astype('float') converts the cleaned strings into floating-point numbers. Finally, dividing by 100 transforms the percentage into a decimal (e.g., 5% becomes 0.05). This step ensures accurate calculations for tax-related computations, improving financial analysis and visualization in the dataset

```python
df['Tax 5%'] = df['Tax 5%'].str.rstrip('%').astype('float') / 100
```

### Total Sales by City

```python
plt.figure(figsize=(8, 5))
city_sales = df.groupby("City")["Total"].sum()
city_sales.plot(kind='bar', color='royalblue')
plt.title("Total Sales by City")
plt.xlabel("City")
plt.ylabel("Total Sales")
plt.show()
```

### This bar chart visualizes total sales across three cities: Mandalay, Naypyitaw, and Yangon. Naypyitaw has the highest sales, while Mandalay and Yangon have nearly equal but slightly lower sales. The balanced sales distribution suggests no extreme variations between cities. Businesses can focus on expanding operations in Naypyitaw to maximize revenue. Lower-performing cities like Mandalay and Yangon may need targeted marketing strategies to boost sales. Discounts, promotions, and improved customer service can help increase engagement. Inventory should be adjusted to match sales demand in each city. Optimizing logistics can prevent stockouts or overstock issues. Local factors affecting sales should be analyzed to adapt pricing and promotions. This data helps in making strategic business decisions for growth.

### Total Sales by Product Line

```python

plt.figure(figsize=(10, 5))
product_sales = df.groupby("Product line")["Total"].sum().sort_values()
product_sales.plot(kind='barh', color='green')
plt.title("Total Sales by Product Line")
plt.xlabel("Total Sales")
plt.ylabel("Product Line")
plt.show()
```

### The **Total Sales by Product Line** chart displays sales across various categories. **Food and beverages** recorded the highest sales, followed closely by **sports and travel**. **Fashion accessories** and **electronic accessories** also performed well. **Home and lifestyle** had competitive sales but ranked slightly lower. **Health and beauty** had the lowest sales among the categories. The sales distribution suggests consumer preference for daily essentials and travel-related products. Electronic and fashion accessories maintain steady demand. Businesses can leverage these insights to focus on high-performing categories. Enhancing marketing for lower-performing segments may improve overall sales balance.

### Sales Trend Over Time

```python

sales_trend = df.groupby("Date")["Total"].sum()
plt.figure(figsize=(12, 5))
sales_trend.plot(color='purple', marker='o', linestyle='-')
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.grid()
plt.show()
```

### The **Sales Trend Over Time** chart shows fluctuations in total sales from January to March 2019. Sales exhibit frequent peaks and dips, indicating inconsistency in daily performance. Several high spikes suggest occasional strong sales days. There are also sharp declines, reflecting periods of low sales. A general upward trend can be observed despite fluctuations. This variability could be influenced by promotions, seasonal demand, or external factors. Identifying patterns in peak sales days could help optimize marketing strategies. Consistent dips may indicate operational inefficiencies or low-demand periods. Businesses should focus on stabilizing sales by analyzing influencing factors. A strategic approach can improve overall revenue growth.

### payment_method_distribution:

```python

plt.figure(figsize=(6, 6))
df['Payment'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title("Payment Method Distribution")
plt.ylabel("")
plt.show()
```

### The **Payment Method Distribution** chart shows the percentage of transactions using different payment methods. E-wallets are the most used, accounting for **34.6%** of transactions. Cash payments follow closely at **34.5%**, showing nearly equal preference with e-wallets. Credit cards make up **31.0%**, indicating a significant but slightly lower usage. The near-equal distribution suggests a diverse customer payment preference. Businesses should ensure seamless support for all payment methods. Encouraging digital payments can enhance transaction efficiency. Promotions on e-wallet or card payments could shift consumer behavior. Understanding payment trends helps in optimizing financial strategies. These insights improve customer experience and business planning.
